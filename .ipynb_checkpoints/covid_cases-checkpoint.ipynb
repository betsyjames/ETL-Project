```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import requests
from sqlalchemy import create_engine
from config import username,password
```

```python
url = 'https://www.worldometers.info/coronavirus/country/us/'

r = requests.get(url)
tables = pd.read_html(r.text)
covid_table = tables[0]
covid_table
```

```python
covid_table = covid_table.drop(columns=['NewCases', 'NewDeaths','TotalTests','Tests/ 1M pop','Population','Source','Projections'])
```

```python
covid_table = covid_table.drop(columns=['#'])
covid_table
```

```python
covid_table = covid_table.drop([59,60,61,62,63])
covid_table
```

```python
covid_table = covid_table.rename(covid_table={"Total Cases/1M pop": "TotalCasespermillpop"})
```

```python
covid_table = covid_table.rename(columns={covid_table.columns[5]: 'totalcasespermillpop'}) 
```

```python
for col in covid_table.columns: 
    print(col)
```

```python
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/etl')
```

```python
engine.table_names()
```

```python
covid_table.to_sql(name='us_covid_case', con=engine, if_exists='append', index=False)
```
