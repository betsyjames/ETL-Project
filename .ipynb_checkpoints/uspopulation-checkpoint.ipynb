```python
import pandas as pd
from sqlalchemy import create_engine 
```

```python
file_to_load = "Resources/2019USPopulation.csv"
population_db = pd.read_csv(file_to_load)
population_db
```

```python
database_path = "Resources/population.sqlite"
engine = create_engine(f"sqlite:///{database_path}") 
```

```python
population_db.to_sql('population', con = engine) 
```

```python
print(engine.execute("SELECT * FROM population").fetchall()) 
```
